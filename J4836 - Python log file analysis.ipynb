Python data analysis of customer log file 'txt' data during DNP load and shot cycles J4836 - CFM-9.4T-VTI-50-H5-DNP

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
DataFrame = pd.read_csv('20211005.txt', delimiter = "\t", names='Time ElapsedTime(sec) SMSField 1stStage 2ndStage Magnet Switch VTIHex HePot Charcoal NVcell NV Pressure'.split())
```

```python
DataFrame.head(3)
```

```python
fig,axes = plt.subplots(1, 2, figsize=(15,5))

axes[0].plot(DataFrame['ElapsedTime(sec)'],DataFrame['VTIHex'],lw=1,color='red')
axes[0].set_title("Plot of VTI Heat Exchanger Over Load and Shot Cycles")
axes[0].axis('tight')
axes[0].set_xlabel('Time / s')
axes[0].set_ylabel('VTI Heat Exchanger T / K')
axes[0].set_ylim([1.4,2])
axes[0].set_xlim([5000,14000])

axes[1].plot(DataFrame['ElapsedTime(sec)'],DataFrame['2ndStage'],lw=1,color='blue')
axes[1].plot(DataFrame['ElapsedTime(sec)'],DataFrame['Magnet'],lw=1,color='green')
axes[1].plot(DataFrame['ElapsedTime(sec)'],DataFrame['HePot'],lw=1,color='red')
axes[1].legend(loc=2)
axes[1].axis('tight')
axes[1].set_xlabel('Time / s')
axes[1].set_ylabel('2nd Stage T / K')
axes[1].set_title("Plot of 2nd Stage Over Load and Shot Cycles")
axes[1].set_ylim([3, 4])
axes[1].set_xlim([5000, 14000])
```

Log file showing VTI heat exchanger temperature following sample loading and shot cycles. Data suggests approximately 1,000 seconds system recovery time between cycles, hence around four cycles possible per hour. Cryocooler 2nd stage, magnet and helium pot temperatures are relatively unpurturbed by the sample loading and shot cycle, remaining below 3.8 K at all times througout the cycle.

```python
fig2,axes2 = plt.subplots(1, 2, figsize=(15,5))

axes2[0].plot(DataFrame['ElapsedTime(sec)'],DataFrame['VTIHex'],lw=1,color='red')
axes2[0].set_title("VTI Heat Exchanger Temperature Over a Single Cycle")
axes2[0].axis('tight')
axes2[0].set_xlabel('Time / s')
axes2[0].set_ylabel('VTI Heat Exchanger T / K')
axes2[0].set_ylim([1.4,2])
axes2[0].set_xlim([8000,9000])

axes2[1].plot(DataFrame['ElapsedTime(sec)'],DataFrame['VTIHex'],lw=1,color='red')
axes2[1].set_title("VTI Heat Exchanger Temperature Over a Single Cycle")
axes2[1].axis('tight')
axes2[1].set_xlabel('Time / s')
axes2[1].set_ylabel('VTI Heat Exchanger T / K')
axes2[1].set_ylim([1.4,2])
axes2[1].set_xlim([11700,12700])
```

Higher scale log files, showing VTI heat exchanger temperature during a single sample loading and shot cycle, indicate that the total time from sample load to base temperature recovery following shot ranges from approximately 500 to 700 seconds. This implies that up to six sample load and shot cycles could be performed each hour.
