```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Import data in tab separated form from txt file and set index column to Month 
data = pd.read_csv('uk_rainfall.txt', sep='\t', index_col='Month')
```

```python
print(data)
```

```python
#Need to remove non-numeric data from rainfall columns 
data['Long term average'] = data['Long term average'].str.replace("mm","").astype(float)
data['Actual rainfall'] = data['Actual rainfall'].str.replace("mm","").astype(float)
```

```python
#Redefine index for plots
index = ['Jan21','Feb21','Mar21','Apr21','May21','Jun21','Jul21','Aug21','Sep21','Oct21','Nov21','Dec21','Jan22','Feb22','Mar22']
data.index = index
print(data)
```

```python
print('Total Average Rainfall: ',data['Long term average'].sum().round(1))
print('Total Actual Rainfall: ',data['Actual rainfall'].sum().round(1))
```

```python
#Plot both rainfall columns on separate axes
fig,ax = plt.subplots(2, 1, sharey=True,figsize=(8,8))
ax[0].plot(data.index,data['Actual rainfall'])
ax[1].plot(data.index,data['Long term average'],color='r')
ax[0].set_ylabel('mm per month')
ax[1].set_ylabel('mm per month')
ax[0].set_title('Actual Rainfall per Month')
ax[1].set_title('Long Term Average Rainfall Per Month')
ax[1].set_xlabel('Month')
plt.tight_layout()
plt.show()
```

```python
#Plot both rainfall columns on the same axes
fig,ax = plt.subplots(figsize=(8,5))
ax.plot(data.index,data['Actual rainfall'],label='Actual')
ax.plot(data.index,data['Long term average'],label='Average',color='r',linestyle='--')
ax.set_xlabel('Month')
ax.set_ylabel('mm per month')
ax.set_title('Actual rainfall per month compared with long term average')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#Use Seaborn to show a box plot of the distribution for both rainfall columns
plt.figure(figsize=(9,6))
sns.boxplot(data=data,palette='Set1').set(ylabel='Monthly rainfall (mm)',title='Box Plot Showing Distribution of Monthly Rainfall Data')
plt.show()
```

The rainfall data for January 2021 to March 2022 looks to be rather out of character when compared with the long term average for each month. Rainfall from May to October 2021 is well above the long term average and falls to below average during the winter period of November 2021 to March 2022.

In addition to this, the box plots show that the distribution of rainfall per month for the period January 2021 to March 2022 is rather more extreme than usual, although by definition we would expect the long term average to smooth out fluctuations in the monthly rainfall over many years. 

The total actual rainfall over the period is approximately 6.4% below average but some months are much wetter than average and some much dryer than average over this period. October 2021 was an extraordinarily wet month with 157.9 mm of rainfall, more than 2.5 times the long term average value for October. It would be interesting to investigate whether the distribution, wetter than average summer and dryer than average winter is indicative of a long term trend or just an anomalous period lacking statistical significance.
