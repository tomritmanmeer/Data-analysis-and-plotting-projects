## Temperature performance of cryogen-free 9 tesla superconducting magnet system following servicing of the Sumitomo pulse tube cryocooler.

Data received in the form of a LabVIEW .dat file extracted from system logging software.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Magnet_Test_9T.dat',sep="\s+")
```

```python
df.head()
```

```python
df.iloc[230]
```

```python
df = df.iloc[230:]
```

Plot system parameters as a function of time to investigate stability.

```python
fig, axs = plt.subplots(1,2,figsize=(10,4))
axs[0].plot(df['Time'],df['B_digital_(T)'], c = 'black')
axs[0].set_title('Magnetic Field Stability - Digital')
axs[0].set_xlabel('Time (s)')
axs[0].set_ylabel('Magnetic Field(T)')

axs[1].plot(df['Time'],df['B_analog_(T)'], c = 'blue')
axs[1].set_title('Magnetic Field Stability - Analogue')
axs[1].set_xlabel('Time (s)')
axs[1].set_ylabel('Magnetic Field (T)')
plt.tight_layout()
```

```python
fig, axs = plt.subplots(1,2,figsize=(10,4))
axs[0].plot(df['Time'],df['1st_Stage'], c='orange')
axs[0].set_title('1st Stage Stability')
axs[0].set_xlabel('Time (s)')
axs[0].set_ylabel('Temperature (K)')

axs[1].plot(df['Time'],df['2nd_Stage'], c='red')
axs[1].set_title('2nd Stage Stability')
axs[1].set_xlabel('Time (s)')
axs[1].set_ylabel('Temperature (K)')
plt.tight_layout()
```

```python
fig, axs = plt.subplots(1,2,figsize=(10,4))
axs[0].plot(df['Time'],df['Magnet'], c='gold')
axs[0].set_title('Magnet Temperature')
axs[0].set_xlabel('Time (s)')
axs[0].set_ylabel('Temperature (K)')

axs[1].plot(df['Time'],df['He_pot'], c='silver')
axs[1].set_title('He Pot Temperature')
axs[1].set_xlabel('Time (s)')
axs[1].set_ylabel('Temperature (K)')
plt.tight_layout()
```

```python
fig, axs = plt.subplots(1,2,figsize=(10,4))

axs[0].plot(df['Time'],df['T_VTI_(K)'], c = 'darkblue')
axs[0].set_title('VTI Temperature (K)')
axs[0].set_xlabel('Time (s)')
axs[0].set_ylabel('Temperature (K)')

axs[1].plot(df['Time'],df['sensor_B_(K)'], c = 'black')
axs[1].set_title('Sample Temperature (K)')
axs[1].set_xlabel('Time (s)')
axs[1].set_ylabel('Temperature (K)')
plt.tight_layout()
```

Distribution plots of magnetic field derived from Hall voltage and temperatures of the cryocooler 2nd stage, magnet and VTI heat exchanger.

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 7))
sns.histplot(df['B_analog_(T)'], kde=True, bins=20,color='blue',ax=axs[0,0])
sns.histplot(df['2nd_Stage'], kde=True, bins=20,color='red',ax=axs[0,1])
sns.histplot(df['Magnet'], kde=True, bins=12,color='gold',ax=axs[1,0])
sns.histplot(df['T_VTI_(K)'], kde=True, bins=25,color='darkblue',ax=axs[1,1])
plt.tight_layout()
```
