```python
#Required imports here
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np
%matplotlib inline
```

```python
#Surface plot as single fiture
fig = plt.figure(figsize=(8,5))
ax = fig.gca(projection='3d')

# Make data.
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)

# Define saddle function z = x**2 - y**2
Z = X**2 - Y**2
# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap=cm.inferno, linewidth=0, antialiased=False)

# Customize the z axis.
ax.set_zlim(-20, 20)
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)
```

```python
fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(121, projection='3d')
ax2 = fig.add_subplot(122, projection='3d')

X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)

R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)
A = (X**2)/4 + (Y**2)/10
# Plot the surface.
surf1 = ax1.plot_surface(X, Y, Z, cmap=cm.viridis, linewidth=0, antialiased=False)
surf2 = ax2.plot_surface(X, Y, A, cmap=cm.plasma, linewidth=0, antialiased=False)
# Customize the z axis.
ax1.set_zlim(-1, 1)
ax1.zaxis.set_major_locator(LinearLocator(10))
ax1.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))
ax2.set_zlim(0, 8)
ax2.zaxis.set_major_locator(LinearLocator(10))
ax2.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

# Add a color bar which maps values to colors.
fig.colorbar(surf1, ax=ax1, shrink=0.5, aspect=5)
fig.colorbar(surf2, ax=ax2, shrink=0.5, aspect=5)
```

```python
x = np.arange(-3,3,0.1)
y = np.arange(-5,5,0.1)
X,Y = np.meshgrid(x,y)
Z = X*np.exp(-X**2 - Y**2)


fig = plt.figure(figsize=(8,5))
ax = fig.add_subplot(111, projection='3d')


# Plot a 3D surface
surface = ax.plot_surface(X, Y, Z, cmap=cm.winter, linewidth=0, antialiased=True)
fig.colorbar(surface, shrink=0.5, aspect=5)
ax.set_zlim(-0.3, 0.3)
ax.set_title('3D Plot')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

```python
fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(121, projection='3d')
ax2 = fig.add_subplot(122, projection='3d')


x = np.arange(-5,5,0.1)
y = np.arange(-5,5,0.1)
X,Y = np.meshgrid(x,y)
Z = X*np.exp(-X**2 - Y**2)


# Plot a basic wireframe
ax1.plot_wireframe(X, Y, Z, rstride=10, cstride=10)
ax1.set_title('row step size 10, column step size 10')


ax2.plot_wireframe(X, Y, Z, rstride=20, cstride=20)
ax2.set_title('row step size 20, column step size 20')
```
